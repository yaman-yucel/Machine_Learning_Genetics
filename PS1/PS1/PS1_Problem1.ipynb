```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
import itertools

# Define the possible outcomes
outcomes = [0, 1]
n = 10
# Generate all combinations of 5 coin flips
combinations = list(itertools.product(outcomes, repeat=n))
```

```python
combinations[-1]
```

```python
def calc_p_MLE(X):
    X_array = np.array(X)
    return np.sum(X_array)/ len(X)
```

```python
print(calc_p_MLE(combinations[3]))
```

```python
def calc_likelihood_given_p(p,X):
    product = 1
    for x in X:
        product *= (p**x * (1-p)**(1-x))
    return product
```

```python
def calc_LRT(combinations):
    LRT_list = []
    
    numerator = calc_likelihood_given_p(1/2,combinations[0])
    for comb in combinations:
        denominator = calc_likelihood_given_p(calc_p_MLE(comb),comb)
        LRT_list.append(numerator/denominator)
    return LRT_list
```

```python
LRT_list = calc_LRT(combinations)
```

```python
a = []
a.append(tuple(np.repeat([1],10)))
p_value = np.where(np.array(LRT_list) <= calc_LRT(a)[0])[0].shape[0] / (2**n)
```
